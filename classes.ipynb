Na poprzednich zajęciach zobaczyliśmy różne sposoby przedstawienia danych w Python, np.:  
- <u> Lista</u>, która umożliwia zebranie <u>danych</u> w jednym miejscu,
- <u> Funkcja</u>, która umożliwia zebranie <u>kodu</u> wykonującego określone zadanie.

Kolejnym sposobem na przedstawienie danych w Pythonie są <u> obiekty</u>, które umożliwiają zebranie funkcji oraz danych w jednym miejscu.

W naszym życiu, czym właściwie są <u>obiekty</u>? Jak je można opisać? Weźmy na przykład piłkę.  
Z piłką możemy wykonać pewne <u>czynności</u>, np:  
- rzucić,
- podnieść,
- kopnąć,
- napełnić powietrzem.  

Możemy też piłkę opisać patrząc na jej <u>cechy</u>, np:
- jaki ma kolor?
- jaki jest jej rozmiar?
- jaka jest jej waga?

Podobnie jest w programowaniu. Jeśli stworzymy obiekt w Python, będziemy mogli nadać mu cechy (takie jak kolor, rozmiar, waga w przypadku piłki), oraz określić czynności które będzie można zastosować (takie jak rzucenie, podniesienie, kopnięcie).

Aby wywołać <u>cechy</u>, które nadaliśmy piłce, napisalibyśmy:  
<i><font color='green'> pilka.kolor  
pilka.rozmiar  
pilka.waga  
</i></font>  

Aby wywołać <u>czynności</u>, które możemy zastosować z piłką, napisalibyśmy:  
<i><font color='green'> pilka.rzuc()  
pilka.podnies()  
pilka.kopnij()
</i></font>

Aby wywołać określone cechy lub czynności, po nazwie obiektu musi nastąpić kropka - jest to sposób zapisu stosowany w wielu językach programowania, nie tylko w Pythonie.

<b> Obiekt = cecha + czynność </b>

<b> Jak stworzyć <u>obiekt</u> w Python?

Tworzenie <u>obiektu</u> w Python jest złożone z 2 etapów:  
1. Określenie jak obiekt będzie wyglądał i jak będzie się zachowywał.  
Można to porównać z projektem budowlanym domu, który powie nam jak dom będzie wyglądał - ale projekt nie jest jeszcze domem. Projekt możemy wykorzystać do wybudowania domu, lub nawet wielu domów.  
2. Stworzenie właściwego obiektu.  
Na podstawie projektu tworzymy właściwy dom.  

Przykład tworzenia klasy:

```python
# Etap 1: Określamy obiekt Piłka
# Nadajemy mu czynność: odbijanie

class Pilka:

    def odbij(self):
        if self.kierunek == 'dół':
            self.kierunek = 'góra'

# Etap 2: Tworzymy właściwy obiekt

mojaPilka = Pilka()
mojaPilka.kierunek = 'dół'
mojaPilka.kolor = 'czerwona'
mojaPilka.rozmiar = 'mała'

print('Właśnie stworzyłem piłkę')
print('Moja piłka jest', mojaPilka.rozmiar)
print('Moja piłka jest', mojaPilka.kolor)
print()
print('Kierunek piłki to', mojaPilka.kierunek)
print('Teraz odbiję piłkę')
mojaPilka.odbij()
print('Teraz kierunek piłki to', mojaPilka.kierunek)
```

Przykład tworzenia klasy z inicjalizacją.  
Inicjalizacja: stworzenie czegoś gotowego do użytku już na starcie

```python
# Etap 1: Określamy obiekt Piłka
# Nadajemy mu czynność: odbijanie

class Pilka:

    def __init__(self, kolor, rozmiar, kierunek):
        self.kolor = kolor
        self.rozmiar = rozmiar
        self.kierunek = kierunek

    def odbij(self):
        if self.kierunek == 'dół':
            self.kierunek = 'góra'

# Etap 2: Tworzymy właściwy obiekt

mojaPilka = Pilka(kolor='czerwona', rozmiar='mała', kierunek='dół')
# mojaPilka.kierunek = 'dół'
# mojaPilka.kolor = 'czerwona'
# mojaPilka.rozmiar = 'mała'

print('Właśnie stworzyłem piłkę')
print('Moja piłka jest', mojaPilka.rozmiar)
print('Moja piłka jest', mojaPilka.kolor)
print()
print('Kierunek piłki to', mojaPilka.kierunek)
print('Teraz odbiję piłkę')
mojaPilka.odbij()
print('Teraz kierunek piłki to', mojaPilka.kierunek)
```
